```python
%%writefile app.py
import streamlit as st
from PIL import Image, ImageFilter, ImageOps
import numpy as np
import cv2

st.title("ðŸŽ¨ Artistik Filtre UygulamasÄ±")

uploaded_file = st.sidebar.file_uploader("Bir resim seÃ§in...", type=["jpg", "jpeg", "png"])

if uploaded_file is not None:
    image = Image.open(uploaded_file)
    st.image(image, caption='Orijinal Resim', use_container_width=True)

    filter_option = st.sidebar.selectbox(
        "Filtre SeÃ§in:",
        ("Yok", "Siyah Beyaz", "BulanÄ±klaÅŸtÄ±r", "Sepia", "Kenar Bulma")
    )

    if filter_option == "Siyah Beyaz":
        processed_img = ImageOps.grayscale(image)
    elif filter_option == "BulanÄ±klaÅŸtÄ±r":
        processed_img = image.filter(ImageFilter.GaussianBlur(5))
    elif filter_option == "Sepia":
        img_array = np.array(image.convert('RGB'))
        img_sepia = cv2.transform(img_array, np.matrix([[0.393, 0.769, 0.189],
                                                        [0.349, 0.686, 0.168],
                                                        [0.272, 0.534, 0.131]]))
        img_sepia = np.clip(img_sepia, 0, 255).astype(np.uint8)
        processed_img = Image.fromarray(img_sepia)
    elif filter_option == "Kenar Bulma":
        processed_img = image.filter(ImageFilter.FIND_EDGES)
    else:
        processed_img = image

    if filter_option != "Yok":
        st.image(processed_img, caption="FiltrelenmiÅŸ Resim", use_container_width=True)
```
