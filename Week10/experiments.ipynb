```python
from transformers import pipeline
import pandas as pd


generator = pipeline("text2text-generation", model="google/flan-t5-base")
```

```python
# Deney parametreleri listesi
test_prompts = [
    "Write a short poem about a robot painting a picture.",
    "Explain quantum physics to a five-year-old."
]

# Farklı yaratıcılık ayarları (Experimenting with parameters)
configs = [
    {"temp": 0.1, "top_p": 0.9, "label": "Düşük Yaratıcılık (Mantıksal)"},
    {"temp": 0.7, "top_p": 0.95, "label": "Orta Yaratıcılık (Dengeli)"},
    {"temp": 1.2, "top_p": 1.0, "label": "Yüksek Yaratıcılık (Rastgele/Yaratıcı)"}
]

results = []

print("--- Deneyler Başlatılıyor ---")
for p in test_prompts:
    for cfg in configs:
        # Metin üretimi
        output = generator(
            p,
            max_length=100,
            temperature=cfg["temp"],
            top_p=cfg["top_p"],
            do_sample=True
        )

        generated_text = output[0]['generated_text']
        results.append({
            "Prompt": p,
            "Ayar": cfg["label"],
            "Temperature": cfg["temp"],
            "Top_p": cfg["top_p"],
            "Sonuç": generated_text
        })

# Sonuçları tablo olarak görselleştir (Record Results)
df = pd.DataFrame(results)
display(df)

# CSV olarak kaydet (README'ye eklemek için kolaylık sağlar)
df.to_csv("experiment_results.csv", index=False)
```
