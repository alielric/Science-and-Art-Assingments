```python
%%writefile app.py
import streamlit as st
from transformers import pipeline

# Model CPU'da √ßalƒ±≈üacak ≈üekilde y√ºkleniyor
chatbot = pipeline("text-generation", model="distilgpt2", device=-1)

st.title("üí¨ Simple Chatbot (Week 6)")

user_input = st.text_input("Enter your message:")

if st.button("Send"):
    response = chatbot(user_input, max_new_tokens=50, truncation=True)
    st.write("**Bot:**", response[0]['generated_text'])
```
